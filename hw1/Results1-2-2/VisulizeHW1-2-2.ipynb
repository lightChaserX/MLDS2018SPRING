```python
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

color = 'rgb'
num_model = 0
for times in range(0,1):
    data_save_file_name = 'mnist'  + '_model' + str(num_model) + '_accu' + str(times) + '.dat'
    loss_save_file_name = 'mnist'  + '_model' + str(num_model) + '_loss' + str(times) + '.dat'
    grad_save_file_name = 'mnist'  + '_model' + str(num_model) + '_grad' + str(times) + '.dat'
    loss_read = pd.read_csv(loss_save_file_name)
    plt.figure(1), 
    plt.semilogy(loss_read, color[times%3], label='Model'+str(num_model))
    plt.xlabel('Number of iterations')
    plt.ylabel('MSE')
    plt.grid(True)
    
    grad_read = pd.read_csv(grad_save_file_name)
    plt.figure(2), 
    plt.semilogy(grad_read, color[times%3], label='Model'+str(num_model))
    plt.xlabel('Number of iterations')
    plt.ylabel('MSE')
    plt.grid(True)


plt.figure(2), plt.legend(loc=0)
plt.figure(2), plt.legend(loc=0)
```
